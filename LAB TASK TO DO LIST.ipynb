```python
class ToDoList:
    def __init__(self):
        self.tasks = []  # list of Tasks

    def add_task(self, task, priority):
        self.tasks.append((task, priority))
        print(f"Task {task} added with priority {priority}")

    def remove_task(self):
        if not self.tasks:
            print("No tasks available to remove!")
            return
        
        # Tasks with the highest priority
        highest_priority = min(self.tasks, key=lambda x: x[1])
        self.tasks.remove(highest_priority)
        print(f"Removed task: '{highest_priority[0]}' (Priority: {highest_priority[1]})")

    def view_tasks(self):
        if not self.tasks:
            print("No tasks in the list!")
        else:
            # Sorting tasks
            sorted_tasks = sorted(self.tasks, key=lambda x: x[1])
            print("To-Do List (Highest Priority on Top):")
            for task, priority in sorted_tasks:
                print(f"   [{priority}] {task}")
            print()

todo = ToDoList()

while True:
    print("Choose an option:")
    print("1.Add Task")
    print("2.Remove Top Priority Task")
    print("3.View Tasks")
    print("4.Exit")
    

    choice = input("Enter your choice: ")
    
    if choice == "1":
        task = input("Enter task: ")
        priority = int(input("Enter priority (1=High, 5=Low): "))
        todo.add_task(task, priority)
    elif choice == "2":
        todo.remove_task()
    elif choice == "3":
        todo.view_tasks()
    elif choice == "4":
        print("Exiting ToDo List. Goodbye!")
        break
    else:
        print("Invalid choice, run again!")
```
